- Extraed los valores √∫nicos su frecuencia. BONUS presenta los resultados en un dataframe. Ver un ejemplo del resultado en apuntes.

```python
def valores_unicos(col):
    dict = {}
    for col in ataques_categoricas.columns:
        dict[col]=ataques_categoricas[col].unique()
    return pd.DataFrame(dict)
```

```python
for col in ataques_categoricas.columns:
    print("------------------------------------------------------------", "\n")
    print(f"En la columna {col} tenemos los siguientes valores unicos")
    print(pd.DataFrame(ataques_categoricas[col].value_counts()))
    print("\n", "------------------------------------------------------------", "\n")
```
