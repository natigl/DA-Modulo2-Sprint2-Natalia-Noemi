```python
#creamos la bbdd 

def crear_bbdd(nombre_bbdd):
    
    mydb = mysql.connector.connect(
      host="localhost",
      user="root",
      password='AlumnaAdalab' 
    )
    print("Conexi칩n realizada con 칠xito")
    
    mycursor = mydb.cursor()

    try:
        mycursor.execute(f"CREATE DATABASE IF NOT EXISTS {nombre_bbdd};")
        print(mycursor)
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

```python
bbdd_ataques= crear_bbdd("tiburones")
```

```python
#creamos la funci칩n para la inserci칩n de las tablas

def insertar_tabla(nombre_bbdd, contrase침a, query):
    
   
    cnx = mysql.connector.connect(user='root', password=f"{contrase침a}",
                                     host='127.0.0.1', database=f"{nombre_bbdd}")
   
    mycursor = cnx.cursor()
    
  
    try: 
        mycursor.execute(query)
        cnx.commit() 
  
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

```python
#creamos la tabla ataques

tabla_ataques = '''
CREATE TABLE IF NOT EXISTS `tiburones`.`ataques` (
  `id_ataques` INT NOT NULL AUTO_INCREMENT,
  `year` INT NOT NULL,
  `type` VARCHAR(250),
  `country` VARCHAR(250),
  `age` FLOAT,
  `species_` VARCHAR(250),
  `fecha_limpia` VARCHAR(250),
  `fatal` VARCHAR(50),
  `sex` VARCHAR(250),
  PRIMARY KEY (`id_ataques`))
ENGINE = InnoDB;
'''
```

```python
#creamos la tabla clima

tabla_clima = '''
CREATE TABLE IF NOT EXISTS `tiburones`.`clima` (
    `id_clima` INT NOT NULL AUTO_INCREMENT,
    `latitud` FLOAT NOT NULL,
    `longitud` FLOAT NOT NULL,
    `country2` VARCHAR(250),
    `fatal_N` INT,
    `fatal_Unknown` INT,
    `fatal_Y` INT, 
    `fatal_N.1` INT,
    `fatal_Unknown.1` INT,
    `fatal_Y.1` INT,
    `species_.1` INT,
    `fecha_limpia.1` INT,
    `type.1` INT,
    `age_NORM` FLOAT NOT NULL, 
 PRIMARY KEY (`id_clima`))
ENGINE = InnoDB
'''
```

```python
insertar_tabla("tiburones", 'AlumnaAdalab', tabla_ataques)
```

```python
insertar_tabla("tiburones", 'AlumnaAdalab', tabla_clima)
```

```python
#insertamos los datos en la tabla clima

for indice, fila in df.iterrows(): 
    
    
    query_clima = f"""
            INSERT INTO clima (latitud, longitud, country2, fatal_N, fatal_Unknown, fatal_Y, fatal_N.1, fatal_Unknown.1, fatal_Y.1, species_.1, fecha_limpia.1, type.1, age_NORM)         
            VALUES ( "{fila['latitud']}", "{fila['longitud']}", "{fila['country2']}", "{fila['fatal_N']}","{fila['fatal_Unknown']}", "{fila['fatal_Y']}","{fila['fatal_N.1']}", "{fila['fatal_Unknown.1']}","{fila['fatal_Y.1']}", "{fila['species_.1']}","{fila['fecha_limpia.1']}", "{fila['type.1']}", "{fila['age_NORM']}");
            """
    
    insertar_tabla("tiburones", 'AlumnaAdalab', query_clima)
```
