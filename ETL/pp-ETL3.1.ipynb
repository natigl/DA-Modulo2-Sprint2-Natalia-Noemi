```python
#creamos la bbdd 

def crear_bbdd(nombre_bbdd):
    
    mydb = mysql.connector.connect(
      host="localhost",
      user="root",
      password='AlumnaAdalab' 
    )
    print("Conexi贸n realizada con 茅xito")
    
    mycursor = mydb.cursor()

    try:
        mycursor.execute(f"CREATE DATABASE IF NOT EXISTS {nombre_bbdd};")
        print(mycursor)
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

```python
#creamos la funci贸n para la inserci贸n de las tablas

def insertar_tabla(nombre_bbdd, contrase帽a, query):
    
   
    cnx = mysql.connector.connect(user='root', password=f"{contrase帽a}",
                                     host='127.0.0.1', database=f"{nombre_bbdd}")
   
    mycursor = cnx.cursor()
    
  
    try: 
        mycursor.execute(query)
        cnx.commit() 
  
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

```python
#creamos la tabla ataques

tabla_ataques = '''CREATE TABLE IF NOT EXISTS `tiburones`.`ataques` (
  `year` INT NOT NULL,
  `type` VARCHAR(250),
  `country` VARCHAR(250),
  `age` FLOAT,
  `species_` VARCHAR(250),
  `fecha_limpia` VARCHAR(250),
  `fatal` VARCHAR(50),
  `sex` VARCHAR(250))
ENGINE = InnoDB; '''
```

```python
#creamos la tabla clima

tabla_clima = ''' CREATE TABLE IF NOT EXISTS `tiburones`.`clima` (
   `latitud` FLOAT,
    `longitud` FLOAT,
    `country2` VARCHAR(250),
    `fatal_N` INT,
    `fatal_Unknown` INT,
    `fatal_Y` INT, 
    `fatal_N.1` INT,
    `fatal_Unknown.1` INT,
    `fatal_Y.1` INT,
    `species_.1` INT,
    `fecha_limpia.1` INT,
    `type.1` INT,
    `age_NORM` FLOAT )
 ENGINE = InnoDB ;'''
```
